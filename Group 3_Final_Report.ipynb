### Methods: Plan

After cleaning the dataset and identifying the two types of music we want to focus on, we aim to examine their association with three emotion levels: Anxiety, Depression, and Insomnia. You can see that the shape and range of all three sets of plots are similar. However, **we cannot conclude that there is no difference between the influence of two genres of music in three different feelings because our comparison is based on only one sample, which cannot generalize to the whole population.**

To address the limitations, we will use the **bootstrap sampling to generate additional samples by resampling our data with replacement**. This approach estimates the sampling distribution without strong assumptions, providing more accurate confidence intervals, especially with smaller sample sizes or non-normal data.

**The significance level (α) is a predefined maximum probability of commuting type I error we’re willing to accept. By default, we set it to be 5%.**

Then, we will **conduct hypothesis tests with a significant level of 5%(by default) to assess the statistical differences in levels of feeling between the genre of Rock music and Pop music**. Specifically, because we don't know the population's standard deviation, we will employ the **two-sample t-test to compare the means of two groups. We will do it three times because we want to focus on the difference of the influence of two music genres in three different feelings.**. Additionally, we will **conduct a 95% confidence interval** to provide a measure of uncertainty, offering stakeholders a range of likely values for the true population parameter.

**Our hypothesis is as follows:** 

Null Hypothesis (H0): There is no discernible difference in the population mean of feeling levels affected by two different genres of music(Rock&Pop).
 
Alternative Hypothesis (H1): A significant difference exists in the population mean of feeling levels affected by two different genres of music(Rock&Pop).

 
That is:
<center>$H_0: \mu_0 = \mu_1$</center> 

<center>$H_1: \mu_0 \neq \mu_1$</center> 


where $\mu$ is the population's mean of the level of each feeling.

The formula of Test statistic we will use:
$$
T = \frac{\bar{x}-\bar{y} - d_0}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}} }
$$

#### Calculating Confidence Interval

```r
# First, we will generate 5000 bootstrap replicates with replacement, then calculate the difference in means between 
#the rank of anxiety, depression and insomnia affected by Pop and Rock music.

# response variable: anxiety level
# exploratory variable: genre
bootstrap_distribution_anxi <- rock_and_pop %>%
                               specify(formula = anxiety ~ genre) %>%
                               generate(reps=5000, type="bootstrap") %>%
                               calculate(stat="diff in means", order = c("Pop", "Rock"))
     
# response variable: depression level
# exploratory variable: genre
bootstrap_distribution_depr <- rock_and_pop %>%
                               specify(formula = depression ~ genre) %>%
                               generate(reps=5000, type="bootstrap") %>%
                               calculate(stat="diff in means", order = c("Pop", "Rock"))

# response variable: insomnia level
# exploratory variable: genre
bootstrap_distribution_inso <- rock_and_pop %>%
                               specify(formula = insomnia ~ genre) %>%
                               generate(reps=5000, type="bootstrap") %>%
                               calculate(stat="diff in means", order = c("Pop", "Rock"))

head(bootstrap_distribution_anxi)
head(bootstrap_distribution_depr)
head(bootstrap_distribution_inso)
```

table 9, 10, 11

```r
# Then, we use a 95% confidence by default for each statistic.
# We have two methods to calculate the lower and upper bound of such a confidence interval.
# 1. Using the percentile method with infer package.
percentile_ci_anxi <- bootstrap_distribution_anxi %>%
                      get_confidence_interval(level=0.95, type="percentile")
percentile_ci_depr <- bootstrap_distribution_depr %>%
                      get_confidence_interval(level=0.95, type="percentile")
percentile_ci_inso <- bootstrap_distribution_inso %>%
                      get_confidence_interval(level=0.95, type="percentile")

percentile_ci_anxi
percentile_ci_depr
percentile_ci_inso
# 2. Using mathematical method(theory based mathod).
# According to the rule of thumb, 95% of values will lie within +- 1.96 standard deviations of the mean.
# The margin of error equals to 1.96*SE, where SE represents the standard error.
conf_ints_anxi <- tibble(n1=point_estimates_mean$n[1],
                         n2=point_estimates_mean$n[2],
                         mu_diff=point_estimates_mean$anxi_mean[1]-point_estimates_mean$anxi_mean[2],
                         SE=sqrt(((point_estimates_sd$anxi_sd[1])^2/n1)+((point_estimates_sd$anxi_sd[2])^2/n2)),
                         MoE=1.96*SE, 
                         lower_ci=mu_diff-MoE,
                         upper_ci=mu_diff+MoE)
conf_ints_depr <- tibble(n1=point_estimates_mean$n[1],
                         n2=point_estimates_mean$n[2],
                         mu_diff=point_estimates_mean$depr_mean[1]-point_estimates_mean$depr_mean[2],
                         SE=sqrt(((point_estimates_sd$depr_sd[1])^2/n1)+((point_estimates_sd$depr_sd[2])^2/n2)),
                         MoE=1.96*SE,
                         lower_ci=mu_diff-MoE,
                         upper_ci=mu_diff+MoE)
conf_ints_inso <- tibble(n1=point_estimates_mean$n[1],
                         n2=point_estimates_mean$n[2],
                         mu_diff=point_estimates_mean$inso_mean[1]-point_estimates_mean$inso_mean[2],
                         SE=sqrt(((point_estimates_sd$inso_sd[1])^2/n1)+((point_estimates_sd$inso_sd[2])^2/n2)),
                         MoE=1.96*SE,
                         lower_ci=mu_diff-MoE,
                         upper_ci=mu_diff+MoE)
conf_ints_anxi
conf_ints_depr
conf_ints_inso
```

### Using the method of Bootstrapping

#### Hypothesis Test:

```r
# First, we will generate 5000 replicates without replacement by setting type="permute".
# Then, we calculate the difference in means between the rank of anxiety, depression and insomnia affected by Pop 
# and Rock music.
# response variable: anxiety level
# exploratory variable: genre
null_distribution_anxi <- rock_and_pop %>% 
                     specify(formula = anxiety ~ genre) %>%
                     hypothesize(null="independence") %>%
                     generate(reps=5000, type="permute") %>%
                     calculate(stat="diff in means", order = c("Pop", "Rock"))
# response variable: depression level
# exploratory variable: genre
null_distribution_depr <- rock_and_pop %>% 
                     specify(formula = depression ~ genre) %>%
                     hypothesize(null="independence") %>%
                     generate(reps=5000, type="permute") %>%
                     calculate(stat="diff in means", order = c("Pop", "Rock"))
# response variable: insomnia level
# exploratory variable: genre
null_distribution_inso <- rock_and_pop %>% 
                     specify(formula = insomnia ~ genre) %>%
                     hypothesize(null="independence") %>%
                     generate(reps=5000, type="permute") %>%
                     calculate(stat="diff in means", order = c("Pop", "Rock"))
head(null_distribution_anxi)
head(null_distribution_depr)
head(null_distribution_inso)
```

### Using the method of Central Limit Theorem(CLT)

#### Calculating test statistics

We will use the following formula to calculate the test statistics for a two-sample t-test.
$$
T = \frac{\bar{x}-\bar{y} - d_0}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}} }
$$

```r
# We have calculate the difference in means and the standard error in the previous part. 
# By using the formula, we can get:

test_statistics_anxi <- conf_ints_anxi$mu_diff/conf_ints_anxi$SE
test_statistics_depr <- conf_ints_depr$mu_diff/conf_ints_depr$SE
test_statistics_inso <- conf_ints_inso$mu_diff/conf_ints_inso$SE

test_statistics_anxi
test_statistics_depr
test_statistics_inso
```

#### Calculating degree of freedom

We will use the following formula to calculate the degree of freedom:

$$
\nu = \frac{
    \left(\frac{s_{Pop}^2}{n_1}+\frac{s_{Rock}^2}{n_2}\right)^2
}
{
\frac{s_{Pop}^4}{n_1^2(n_1-1)}+\frac{s_{Rock}^2}{n_2^2(n_2-1)}
}
$$

```r
degree_of_freedom_anxi <- (conf_ints_anxi$SE)^4/
                     (((point_estimates_sd$anxi_sd[1]^4)/
                     ((point_estimates_sd$n[1])^2*(point_estimates_sd$n[1]-1)))+
                     (point_estimates_sd$anxi_sd[2]^4)/
                     (((point_estimates_sd$n[2])^2*(point_estimates_sd$n[2]-2))))

degree_of_freedom_depr <- (conf_ints_depr$SE)^4/
                     (((point_estimates_sd$depr_sd[1]^4)/
                     ((point_estimates_sd$n[1])^2*(point_estimates_sd$n[1]-1)))+
                     (point_estimates_sd$depr_sd[2]^4)/
                     (((point_estimates_sd$n[2])^2*(point_estimates_sd$n[2]-2))))

degree_of_freedom_inso <- (conf_ints_inso$SE)^4/
                     (((point_estimates_sd$inso_sd[1]^4)/
                     ((point_estimates_sd$n[1])^2*(point_estimates_sd$n[1]-1)))+
                     (point_estimates_sd$inso_sd[2]^4)/
                     (((point_estimates_sd$n[2])^2*(point_estimates_sd$n[2]-2))))

degree_of_freedom_anxi
degree_of_freedom_depr
degree_of_freedom_inso
```

#### Calculating P-value

```r
# Because our three test statistics are all negative, we want to use pt to calculate the area to the left of the 
# test statistics without the code "lower.tail=FALSE".
# Because we are doing a two sided hypothesis test, we need to multiple the result of pt function by 2.

p_value_anxi <- 2*pt(test_statistics_anxi, df=degree_of_freedom_anxi)
p_value_depr <- 2*pt(test_statistics_depr, df=degree_of_freedom_depr)
p_value_inso <- 2*pt(test_statistics_inso, df=degree_of_freedom_inso)

p_value_anxi
p_value_depr
p_value_inso
```

#### Making Conclusion of Hypothesis Test

Because we use a 5% significance level, alpha=0.05. On the one hand, if the p-value is lower than 0.05, we will reject H0 and prefer H1. On the other hand, if the p-value is greater or equal to 0.05, we fail to reject H0.
As a memory:
<center>$H_0: \mu_0 = \mu_1$</center> 

<center>$H_1: \mu_0 \neq \mu_1$</center> 
We can make three conclusions based on the method of CLT:

For the level of the **anxiety**, because 0.8764>0.05, we **fail to reject the null hypothesis** that Pop music and Rock music have an equal impact on the level of anxiety.

For the level of the **depression**, because 0.0271<0.05, we **reject the null hypothesis and prefer the alternative hypothesis** that Pop music and Rock music have distinct effects on the level of depression.

For the level of the **insomnia**, because 0.1428>0.05, we **fail to reject the null hypothesis** that Pop music and Rock music have an equal impact on the level of insomnia.
